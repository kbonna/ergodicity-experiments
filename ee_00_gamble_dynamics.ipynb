```python
import numpy as np
import matplotlib.pyplot as plt
```

### Ergodic property
$$\lim_{\Delta t\rightarrow\infty}\frac{1}{\Delta t}\int_{t}^{t+\Delta t}A(s)\,ds=\lim_{N\rightarrow\infty}\frac{1}{N}\sum_{i}^{N}A_i$$

### Gamble
A gamble is a pair of a random variable, $D$, and a duration, $\delta t$.  

$D$ is called the payout and takes one of $N$ (mutually exclusive) possible monetary values, $\{D_1, \dots, D_N\}$, associated with probabilities, $\{p_1, \dots, p_N\}$, where $\sum_{i=1}^{N}p_i=1$. Payouts can be positive, associated with a monetary gain, or negative associated with a loss. We order them such that $D_1 < \dots < D_N$.

### Additive repetition
$$\delta x(t)\equiv x(t+\delta t)-x(t)$$
$$\delta x(t)=D$$
Wealth after $T$ rounds:
$$x(t_0+T\delta t)=x(t_0)+\sum_{\tau =1}^{T}D(\tau)$$

```python
d = np.array([10, -10])
p = np.array([0.5, 0.5])

np.random.seed(0)
x = np.cumsum(np.random.choice(d, size=100, p=p))
plt.plot(x)
plt.axhline(0, color="k", linestyle=":")
```

### Multiplicative repetition
Payout is expressed as a random wealth multiplier
$$r \equiv \frac{x(t_0)+D}{x(t_0)}$$
$$x(t+\delta t)=rx(t)$$
Successive changes in wealt are not stationary as they depend on $t$ through $x(t)$:
$$\delta x(t)=(r-1)x(t)$$
Wealth after $T$ rounds:
$$x(t_0+T\delta t)=x(t_0)\prod_{\tau =1}^{T}r(\tau)$$

```python
d = np.array([1.5, 0.6])
p = np.array([0.5, 0.5])

np.random.seed(1)
x0 = 1 # initial money
x = np.cumprod(np.random.choice(d, size=100, p=p)) * x0 
plt.plot(x)
plt.axhline(x0, color="k", linestyle=":")
```
