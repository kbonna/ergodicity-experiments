```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
np.random.seed(0)
```

```python
plt.style.use("ggplot")
mpl.rcParams.update({
    "font.size": 15,
    "figure.figsize": (7, 4.5)
})
```

# Additive dynamics

```python
d = np.array([1.5, -.5])
p = np.array([0.5, 0.5])

x = np.cumsum(np.random.choice(d, size=100_000, p=p))
plt.plot(x)
plt.ylabel("$x$")
plt.xlabel("$t$")
plt.title("Additive dynamics")
```

# Multiplicative dynamics

```python
d = np.array([2, 0.6])
p = np.array([0.5, 0.5])

x = np.cumprod(np.random.choice(d, size=100_000, p=p))
plt.plot(np.log(x))
plt.ylabel("$\log{x}$")
plt.xlabel("$t$")
plt.title("Multiplicative dynamics")
```

# Square root dynamics

**Example**
- current wealth: $x(t) = 100$
- gamble: with probability $\frac{1}{2}$ you gain $1.2\sqrt{x(t)}=12$ or you lose $1\sqrt{x(t)}=10$


```python
d = np.array([1.2, -1])
p = np.array([0.5, 0.5])

x = np.random.choice(d, size=100_000, p=p)
x[0] = 100

for i in range(1, len(x)):
    x[i] = x[i-1] + x[i] * np.sqrt(x[i-1]) 

plt.plot(np.sqrt(x))
plt.ylabel("$\sqrt{x}$")
plt.xlabel("$t$")
plt.title("Square root dynamics")
```
